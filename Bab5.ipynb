Ada empat perintah utama dalam DML:

1. INSERT â€“ memasukkan data baru ke tabel
2. SELECT â€“ menampilkan data
3. UPDATE â€“ mengubah data
4. DELETE â€“ menghapus data

Semua perintah ini berkaitan dengan isi tabel, bukan strukturnya.

1. INSERT (Memasukkan Data)

Insert digunakan untuk menambahkan baris baru ke tabel.

Ada 2 cara:

a. INSERT tanpa menyebut nama kolom

INSERT INTO nama_tabel
VALUES (isi1, isi2, isi3, ...);

b. INSERT dengan menyebut nama kolom

INSERT INTO nama_tabel (kolom1, kolom2, kolom3)
VALUES ('isi1', 'isi2', 'isi3');

Kelebihan: lebih aman karena tidak perlu mengisi semua kolom.

2. SELECT (Menampilkan Data)

Digunakan untuk melihat isi tabel.

a. Menampilkan semua data

SELECT * FROM nama_tabel;

b. Menampilkan kolom tertentu

SELECT kolom1, kolom2 FROM nama_tabel;

c. Menampilkan baris tertentu (menggunakan WHERE)

SELECT * FROM nama_tabel
WHERE kolom = 'nilai';

3. UPDATE (Mengubah Data)

Digunakan untuk memperbarui data pada baris tertentu.

Format:

UPDATE nama_tabel
SET kolom = 'nilai_baru'
WHERE kolom_kunci = 'nilai';

> Sangat penting: Gunakan WHERE, kalau tidak semua data akan ikut berubah.

4. DELETE (Menghapus Data)

Menghapus baris tertentu dari tabel.

Format:

DELETE FROM nama_tabel
WHERE kolom_kunci = 'nilai';

> Sama seperti UPDATE, WAJIB pakai WHERE, kalau tidak semua data terhapus.


5. SELECT Dengan Kondisi Tambahan (Operator Relasional)

Operator untuk membandingkan nilai:

Operator    Arti

=   sama dengan
>   lebih besar
<   lebih kecil
>=  lebih besar sama dengan
<=  lebih kecil sama dengan
<>  tidak sama dengan


Contoh:

SELECT * FROM products
WHERE prod_price >= 20;
